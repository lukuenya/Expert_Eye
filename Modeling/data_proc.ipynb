```python
import pandas as pd
import numpy as np
data8 = pd.read_excel('P:\DATA_OCT_22\Expert_Eye\Dataset\Data\data_v11.xlsx')
data8.head()
```

```python
# merge data8 with data1_hads
data8 = pd.merge(data8, data1_hads, on='Foldername', how='left')
data8.head()
```

```python
# Create a grip column in data8 = max(grip_left_hand, grip_right_hand)
data8['grip'] = data8[['grip_left_hand', 'grip_right_hand']].max(axis=1)
data8.head()
```

```python
# drop grip_left_hand and grip_right_hand
data8.drop(['grip_left_hand', 'grip_right_hand'], axis=1, inplace=True)
data8.head()
```

```python
# create walking_speed_4m column in data8 = max(SPPB06, SPPB07)
data8['walk_time_4m'] = data8[['SPPB06', 'SPPB7']].max(axis=1)
data8.head()
```

```python
# drop SPPB06 and SPPB07
data8.drop(['SPPB06', 'SPPB7'], axis=1, inplace=True)
```

```python
#drop the column Weight_Diff
data8.drop(['Weight_Diff'], axis=1, inplace=True)
```

```python
# Weight_Diff = (Past_Weight - Weight)/Past_Weight
data8['Weight_Diff'] = (data8['Past_Weight'] - data8['Weight'])/data8['Past_Weight']
```

```python
data8['Weight_Diff'].head(50)
```

```python
# For Item_1
data8['Item_1'] = data8['Weight_Diff'].apply(lambda x: 1 if x <= -3 else (0 if x > -3 else np.nan))
data8.head()
```

```python
# For Item_2
data8['Item_2'] = data8['HADS_D_Score'].apply(lambda x: 1 if x > 10 else (0 if x <= 10 else np.nan))
data8.head()
```

```python
# Define Item_3
data8['Item_3'] = data8['EXAMCLIN02'].apply(lambda x: 0 if x == 1 else (1 if x == 0 else np.nan))
data8.head()
```

```python
# Item 4 
def calculate_item_4(row):
    if pd.isna(row['Gender']) or pd.isna(row['Height']) or pd.isna(row['walk_time_4m']):
        return np.nan

    if row['Gender'] == 1:
        if row['Height'] <= 1.73:
            cutoff = 7
        else:
            cutoff = 6
    elif row['Gender'] == 0:
        if row['Height'] <= 1.59:
            cutoff = 7
        else:
            cutoff = 6
    
    return 1 if row['walk_time_4m'] >= cutoff else 0

data8['Item_4'] = data8.apply(calculate_item_4, axis=1)
data8.head()
```

```python
# Item 5
import numpy as np
def calculate_item_5(row):
    if pd.isna(row['Gender']) or pd.isna(row['BMI']) or pd.isna(row['grip']):
        return np.nan

    if row['Gender'] == 1:
        if row['BMI'] <= 24:
            cutoff = 29
        elif 24.1 <= row['BMI'] <= 26:
            cutoff = 30
        elif 24.1 <= row['BMI'] <= 28:
            cutoff = 30
        else:                       # row['BMI'] > 28
            cutoff = 32
    else:                           # Gender == 0
        if row['BMI'] <= 23:
            cutoff = 17
        elif 23.1 <= row['BMI'] <= 26:
            cutoff = 17.3
        elif 26.1 <= row['BMI'] <= 29:
            cutoff = 18
        else:                       # row['BMI'] > 29
            cutoff = 21

    return 1 if row['grip'] > cutoff else 0

data8['Item_5'] = data8.apply(calculate_item_5, axis=1)
data8.head()
```

```python
# 'Fried_Score' is the sum of the 5 items (Item_1, Item_2, Item_3, Item_4, Item_5). If one of the items is NaN, then the 'Fried_Score' is the sum of the other items.
data8['Fried_Score'] = data8[['Item_1', 'Item_2', 'Item_3', 'Item_4', 'Item_5']].sum(axis=1)
data8.head()
```

```python
# 'Fried_State' is 1 if 'Fried_Score' >= 3, else 0
data8['Fried_State'] = data8['Fried_Score'].apply(lambda x: 1 if x >= 3 else 0)
data8.head()
```
